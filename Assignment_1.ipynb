## TASK 1 - Simple ATM Calculator

Write a program that takes the user's account balance and withdrawal amount. If
sufficient balance is available, deduct the amount and display the updated balance.
Otherwise, print an error message.

Sample Input:

Balance = 5000

Withdrawal = 1500

Sample Output:

Transaction successful. Remaining balance: ₹3500

```python
# Taking account balance from the user and converting it to an integer
balance = int(input('Enter your account balance: ₹'))

# Taking withdrawal amount from the user and converting it to an integer
withdrawal = int(input('Enter ypur withdrawal amount: ₹'))

# Checking if the withdrawal amount is less than or equal to the balance amount
if withdrawal <= balance:

  # deducting the withdrawal amount from the balance
  balance -= withdrawal

  # printing a success message along with the updated balance
  print('Transaction successful. Remaining balance: ₹',balance)

# if the withdrawal amount is more than the balance amount
else:

  #printing an error message for insufficient balance.
  print('Insufficient balance.')
```

## TASK 2 - Electricity Bill Generator

Create a bill calculator where the user enters the number of units consumed. The rate is
₹5 per unit for the first 100 units, ₹7 for the next 100, and ₹10 for anything above 200
units. Calculate and display the total bill.

Sample Input:

Units consumed = 275

Sample Output:
\
Bill amount = ₹1950 (100×5 + 100×7 + 75×10)

```python
# Taking input from the user for units consumed
unit = int(input('Enter the number of units consumed: '))

# Initializing the bill amount to 0
bill=0

# checking if the units consumed are less than or equal to 100
if unit<=100:

  bill = unit * 5 # if units are 100 or less, charge ₹5 per unit

# Checking if the unit consumed are between 101 and 200
elif unit<=200:

  # Charge ₹5 for first 100 units
  # Charge ₹7 for the remaining units (units - 100)
  bill = (100*5) + (unit-100)*7

# If the unit consumed is more than 200
else:

  # Charge ₹5 for first 100 units
  # Charge ₹7 for the next units
  # Charge ₹10 for the remaining units (units - 200)
  bill = (100*5) + (100*7) + (unit-200)*10

# Print the total calculated bill
print(f'Bill amount = ₹{bill}')
```

# TASK 3 - Temperature Converter
Write a Python program to convert a given temperature value between Celsius and
Fahrenheit. Accept input as a number and a unit indicator ('C' or 'F').

Sample Input:
\
Temp = "37"
\
Unit = "C"

Sample Output:
\
Temperature in Fahrenheit: 98.6°F

```python
# Taking temperature input from the user
temp = int(input('Enter the temperature value: '))

# Taking unit input (C or F) from the user
unit = input('Enter the unit (C for Celsius /F for Fahrenheit): ').upper()

# checking the unit and converting accordingly
if unit == 'C':
  # If input is in Celsius, convert to Fahrenheit using formula
  fahrenheit = (temp * 9/5) + 32

  # Displaying the converted temperature in Fahrenheit
  print(f'Temperature in Fahrenheit: {fahrenheit}°F')

elif unit == 'F':

  # If input is in Fahrenheit, convert to Celsius using formula
  celsius = (temp - 32) * 5/9

  # Displaying the converted temperature in Celsius
  print(f'Temperature in Celsius: {celsius}°C')

else:
  # If the unit is not C or F, print an error message
  print('Invalid unit. Please enter \'C\' for Celsius or \'F\' for Fahrenheit.')

```

# Task 4 - Mobile Recharge Validity

Write a program that takes the recharge date and today’s date as input and checks
whether the 28-day recharge plan is still valid or expired.

Sample Input:
\
Recharge date: 2025-07-01
\
Today : 2025-07-30


Sample Output:
\
Plan expired. Recharge again.

```python
# Importing datetime module
# We use this to convert date strings into date objects for comparison
from datetime import datetime

# Asking the user to enter the recharge date in YYYY-MM-DD format
recharge_date_str = input('Enter the recharge date (YYYY-MM-DD):')

# Asking the user to enter today's date in the same format
today_str = input('Enter today\'s date (YYYY-MM-DD):')

# Converting the entered recharge date string into a proper date format
recharge_date = datetime.strptime(recharge_date_str, '%Y-%m-%d').date()

# Converting today's date string into a proper date format
today = datetime.strptime(today_str, '%Y-%m-%d').date()

# Calculating how many days have passed since the recharge date
days_passed = (today - recharge_date).days

# Checking if the plan is still valid (within 28 days)
if days_passed <= 28:
    # Displaying how many days are left before the 28-day plan expires
    print(f'Plan is still valid. {28 - days_passed} day(s) remaining.')

else:
    # Displaying that the 28-day validity is over and they need to recharge again
    print('Plan expired. Recharge again.')
```

# TASK 5 - Bus Seat Reservation System

A list contains the names of passengers in order of booking. A cancellation occurs; remove
the name from the list and shift the rest. Also, insert a new booking at a specified position.

Sample Input:
\
Booked = ['Amit', 'Sara', 'John'],
\
Cancel = 'Sara',
\
New = 'Neha', Position = 1

Sample Output:
\
['Amit', 'Neha', 'John']

```python
# Taking initial list of booked passengers as input from the user
booked_input = input("Enter the list of booked passengers (comma-separated): ")

# Converting the input string into a list usinf comma as seperator
booked = booked_input.split(',')

# Looping through each index in the booked list
for i in range(len(booked)):
  # Removing leading/trailing spaces from each name
  booked[i] = booked[i].strip()


# Asking user for the name of the passenger who wants to cancel the booking
cancel = input("Enter the name of the passenger to cancel: ").strip()

# Check if the name to be canceled exists in the booked list
if cancel in booked:
  # if the name is found in the list, then remove the name from the list
  booked.remove(cancel)
else:
  #if the name is not found in the list, then show message.
  print(f"{cancel} not found in the list.")

# Asking user for the new passenger's name
new_name = input("Enter the name of the new passenger: ").strip()

# Asking user at which position (index) the new passenger should be inserted
position = int(input("Enter the position to insert the new passenger (starting from 0): "))

# Inserting the new passenger's name at the specified position
booked.insert(position, new_name)

# Displaying the final updated list of booked passengers
print("Updated passenger list:", booked)
```

# TASK 6 - Sum of Even Numbers

Write a program that takes a number as input and prints the sum of all even numbers
from 1 to that number.

Sample Input:\
Number: 10

Sample Output:\
Sum of even numbers: 30\
(2 + 4 + 6 + 8 + 10)

```python
# Asking the user to input an integer number
num_value = int(input('Enter a number: '))

# Initializing the sum of even numbers to 0
sum_even = 0

# Looping from 1 to the user-entered number(inclusive)
for num in range(1, num_value+1):
  # Checking if the current number is even
  if num % 2 ==0:
    # If it is True, then adding the even number to the sum
    sum_even += num

# Printing the final sum of even numbers
print(f'Sum of even numbers from 1 to {num_value} : {sum_even}')
```

# TASK 7 - Dictionary – Student Marks Summary

Write a program that takes input of 3 students’ names and their marks, stores them in a
dictionary, and prints the average marks.

Sample Input:\
Enter name: Alice\
Enter marks: 80 \
Enter name: Bob \
Enter marks: 90 \
Enter name: Charlie \
Enter marks: 70

Sample Output:
Average Marks: 80.0

```python
# Creating an empty dictionary to store student names as keys and marks as values
student_marks = {}

# Looping 3 times to take input for 3 students
for i in range(3):
  # Taking student's name as input
  name = input('Enter name: ')
  # Taking student's marks as input and converting it to integer
  marks = int(input('Enter marks: '))
  # Adding the name and marks to the dictionary
  student_marks[name] = marks

# Calculating the total marks by adding all the values in the dictionary
total_marks = sum(student_marks.values())

# Calculating average by dividing total marks by number of students
average_marks = total_marks / len(student_marks)

# Printing the average marks
print(f'Average Marks: {average_marks}')
```

# TASK 8 - Dictionary – Word Frequency

Write a program that takes a sentence and displays the frequency of each word using a
dictionary.

Sample Input:\
Sentence: hello world hello python

Sample Output:\
hello: 2 \
world: 1 \
python: 1

```python
# Taking a sentence as input from the user
sentence = input("Enter a sentence: ")

# Splitting the sentence into words using split() function
words = sentence.split()

# Creating an empty dictionary to store word frequencies
word_freq = {}

# Looping through each word in the list of words
for word in words:
    # If the word is already in the dictionary, increase its count by 1
    if word in word_freq:
        word_freq[word] += 1
    # If the word is not in the dictionary, add it with value 1
    else:
        word_freq[word] = 1

# Displaying the frequency of each word
for word, freq in word_freq.items():
    print(f"{word}: {freq}")
```

# TASK 9 - Functions – Check Prime

Write a function that checks whether a number is prime. Call the function with user input.

Sample Input:\
Enter a number : 17

Sample output:\
17 is a prime number.

```python
# Function to check if a number is prime
def is_prime(num_val):
  # Prime numbers are greater than 1
  if num_val <= 1:
    return f'{num_val} is not a prime number'

  # Loop from 2 to one less than the number (not including num_val itself)
  for num in range(2, num_val):
    # If the number is divisible by any num in this range, it's not prime
    if num_val % num == 0:
      return f'{num_val} is not a prime number'

  # If loop completes with no divisors found, it's a prime
  return f'{num_val} is a prime number'
```

```python
# Take input from user and convert to integer
number = int(input('Enter a number: '))

# Call the function and print the result
print(is_prime(number))
```

# TASK 10 - Vowel Counter

Write a function that takes a string and returns the number of vowels in it.

Sample Input:\
Enter a string: Python Programming

Sample output:\
Number of vowels: 4

```python
# Define a function named vowel_count that takes one argument: text
def vowel_count(text):

  # Create a string containing all vowels (both lowercase and uppercase)
  vowels = 'aeiouAEIOU'

  # Initialize the count of vowels to 0
  count = 0

  # Loop through each character in the input text
  for ch in text:

    # Check if the character is in the vowels string
    if ch in vowels:
      count += 1   # If it's a vowel, increment the count by 1

  # Return the total number of vowels found in the text
  return f'No of Vowels: {count}'
```

```python
# Take a string input from the user
text_value = input('Enter a string: ')

# Call the vowel_count function with user input and print the result
print(vowel_count(text_value))
```

# TASK 11 - . Count Divisible Numbers

Write a program that takes two numbers as input: a range limit N and a divisor D. Count
how many numbers between 1 and N are divisible by D.

Sample Input:\
Enter the range limit (N): 50 \
Enter the divisor (D): 5

Sample output:\
Count of numbers divisible by 5: 10\
(i.e., 5, 10, 15, ..., 50)

```python
# Taking input from the user for the upper limit of the range (N)
limit = int(input('Enter the range limit (N): '))

# Taking input from the user for the divisor (D)
divisor = int(input('Enter the divisor (D): '))

# Initializing a counter to keep track of how many numbers are divisible by D
count = 0

# Looping through all numbers from 1 to N (inclusive)
for num in range(1, limit + 1):
  # Checking if the current number is divisible by the divisor
  if num % divisor == 0:
    # If divisible, increment the counter
    count += 1

# Printing the total count of numbers divisible by the divisor
print(f'Count of numbers divisible by {divisor}: {count}')
```
