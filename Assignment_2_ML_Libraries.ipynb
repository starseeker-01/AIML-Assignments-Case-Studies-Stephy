# **NumPy (Array Operations)**


## TASK 1

Create a NumPy array of shape (5, 5) with random integers between 1 and 50.

*  Find the row with the maximum sum.
*  Replace all elements greater than the mean with 0



```python
# Import NumPy library for numerical operations and arrays
import numpy as np

# Create a 5x5 NumPy array with random integers from 1 to 50
# (inclusive of 1, exclusive of 51)
arr = np.random.randint(1, 51, size=(5, 5))

#Print the generated array
print('Original Array:\n',arr)

# Find the row with the maximum sum

# Calculate the sum of each row (axis=1 means sum across columns)
row_sum = np.sum(arr, axis=1)

# Find the index of the riow with the maximum sum
max_row_index = np.argmax(row_sum)

# Print the sum of each row
print('\n Row Sums:',row_sum)
# Print the index of the row with the maximum sum
print('\n Row with maximum sum (index):',max_row_index)
# Print the row itself that has the maximum sum
print('\n Row with maximum sum:', arr[max_row_index])

# Replace all elements greater than the mean with 0

# Calculate the mean (average) value of all elements in the array
mean_value = np.mean(arr)

# Loop through each row (i) of the array
for i in range(arr.shape[0]):
  # Loop through each column (j) of the array
  for j in range(arr.shape[1]):
    # If the element is greater than the mean value
    if arr[i][j] > mean_value:
      arr[i][j] = 0 # set it to 0

# Print the mean va
print('\n Mean Value:', mean_value)

# Print the updated array after replacing values greater than the mean with 0
print('\n Updated Array:\n',arr)
```

# **Pandas (Data Handling)**

## TASK 3
Load the dataset Iris data

* Display the first 10 rows and basic statistics.
* Filter all rows where sepal_length > median and species is "versicolor".
* Group the data by species and compute the mean petal_length.





```python
# Importing the necessary libraries
import pandas as pd # For data handling

from sklearn.datasets import load_iris # To load the Iris dataset

# 1. Load the Iris dataset
iris = load_iris()  # Load Iris dataset from sklearn

# Create DataFrame with feature names as column headers
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add species column using target values mapped to target names
df['species'] = iris.target_names[iris.target]

# 2. Display the first 10 rows of the dataset
print('First 10 rows of the dataset:')
print(df.head(10))

print('\n')

# 3. Display basic statistics
print('\n Basic Statistics of the dataset:')

# info()
# Displays basic information about the dataset,
# Shows column names, data types, non-null values, and memory usage
print(df.info())

print('\n')

# describe()
# Displays basic statistics for numerical columns
# Includes: count, mean, standard deviation, minimum, quartiles, and maximum values
print(df.describe())
```

```python
# 4. Filter rows where 'sepal length (cm)' > median AND species is 'versicolor'
print("\nFiltered rows (sepal length > median & species = versicolor):")

# Calculate the median value of the 'sepal_length' column
median_value = df['sepal length (cm)'].median()

# Filter the DataFrame to include only rows where:
# 1. 'sepal_length' is greater than the calculated median value
# 2. 'species' is equal to 'versicolor'
filtered_df = df[(df['sepal length (cm)'] > median_value) & (df['species'] == 'versicolor')]

# Display the filtered DataFrame
print(filtered_df)
```

```python
# 5. Group the data by 'species' and compute mean of 'petal_length'
grouped_mean_petal_length = df.groupby('species')['petal length (cm)'].mean()

# Display the mean petal length for each species
print("\nMean petal length for each species:")
print(grouped_mean_petal_length)
```

## Task 4

Add a new column named petal_ratio = petal_length / petal_width.
* Find the top 5 rows with the highest
petal_ratio.

```python
# Add a new column 'petal_ratio' which is calculated as petal_length / petal_width
df['petal_ratio'] = df['petal length (cm)'] / df['petal width (cm)']

# Find the top 5 rows with the highest 'petal_ratio' values
top_5_petal_ratio = df.nlargest(5, 'petal_ratio')

# Print the top 5 rows with the highest petal_ratio
print("Top 5 rows with the highest petal_ratio:")
print(top_5_petal_ratio)
```

# **Data Visualization**

## TASK 5

Using Matplotlib, create:
* A histogram of sepal_length with 15 bins.
* A scatter plot of sepal_length vs petal_length, color-coded by species.

```python
# Importing matplotlib for data visualization
import matplotlib.pyplot as plt
```

### Histogram of sepal length with 15 bins

```python
# Creating a histogram of sepal length with 15 bins
# filled with cyan color and blue edges for better visibility
plt.hist(df['sepal length (cm)'], bins=15, color='cyan', edgecolor='blue')

# Set the title of the histogram plot
plt.title('Histogram of Sepal Length')

# Label the x-axis as 'Sepal Length'
plt.xlabel('Sepal Length')

# Label the y-axis as 'Frequency'
plt.ylabel('Frequency')

# Display the histogram plot on the screen
plt.show()
```

**Insights:**

This histogram displays the frequency of distribution of Sepal Length in the dataset. Most flowers have sepal lengths between 5.0–6.5 cm, with peaks around 5.0 cm and 5.5 cm (about 20 samples each). Very short (<4.8 cm) and very long (>7.2 cm) sepals are uncommon, indicating possible outliers. The distribution is generally bell-shaped with a slight right skew, as there are fewer flowers with larger sepal lengths compared to smaller ones. Overall, the histogram shows that the majority of the sepal lengths cluster in the mid-range, with a steady decrease in frequency as sepal length increases.

### Scatter plot of sepal_length vs petal_length, color-coded by species


```python
# Create a scatter plot of Sepal Length vs Petal Length
# 'c' assigns colors to points based on species, mapped to numeric values (0, 1, 2) for plotting
plt.scatter(df['sepal length (cm)'],
            df['petal length (cm)'],
            c=df['species'].map({'setosa': 0, 'versicolor': 1, 'virginica': 2}))

# Set the label for the X-axis
plt.xlabel('Sepal Length')

# Set the label for the Y-axis
plt.ylabel('Petal Length')

# Set the title of the scatter plot
plt.title('Scatter Plot of Sepal Length vs Petal Length (Color-coded by Species)')

# Display the plot on the screen
plt.show()
```

**Insight:**

The scatter plot shows a clear positive correlation between sepal length and petal length across all species, forming three distinct clusters. Setosa(purple) has consistently shorter petals and smaller sepals, forms a distinct low-range cluster ie, easily identifiable.  Versicolor(teal) occupies the mid-rangewith a clear positive trend but partially overlaps with Virginica (yellow), which has the largest sepal and petal lengths. The overall patterm highlights strong species separation for Setosa, partial overlap between Versicolor and Virginica, and a general positive relationship between sepal and petal lengths.

Using Seaborn, plot:

### Pairplot

```python
# Importing seaborn for data visualization
import seaborn as sns

# Create a pairplot for the dataset
sns.pairplot(df)

## Displaying the pairplot
plt.show()
```

**Insights:**

The pairplot of the Iris dataset shows that petal length and petal width have a strong positive correlation and clearly separate one species from the others. Sepal length and sepal width display weaker relationships, with more scattered distributions. The diagonal histograms indicate skewed distributions for petal features and a near-normal distribution for sepal width. While one species forms a distinct cluster, two species overlap in the mid-range values, especially when considering only sepal features.

#### Boxplot

```python
# Creates a boxplot of petal width for each species with different colors
sns.boxplot(x='species', y='petal width (cm)', data=df, hue='species', palette='cool')

# Adds a title to the boxplot
plt.title('Boxplot of Petal Width by Species')

# Label the x-axis as 'Species'
plt.xlabel('Species')

# Label the y-axis as 'Petal Width'
plt.ylabel('Petal Width')

# Displays the plot
plt.show()
```

**Insights:**

The boxplot of petal width by species shows clear differences among the three Iris species. *Setosa* has the smallest petal width, with a very low median around 0.2 cm and minimal variation, although a couple of outliers are present. *Versicolor* has a higher median of about 1.3 cm with moderate variability, while *Virginica* exhibits the largest petal widths, with a median around 2.0 cm and the widest spread among the three. The distinct separation of ranges between species, with little to no overlap, indicates that petal width is a strong distinguishing feature for classifying Iris species.
